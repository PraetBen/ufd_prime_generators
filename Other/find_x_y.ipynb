```python
def find_c_d(to_find):

    c_limit = math.ceil(math.sqrt(to_find))
    d_limit = math.ceil(math.sqrt(4*to_find/163))
    c = 0
    d = 1

    found_pairs = []
    absVal = get_absolute_value(c,d)
    absValNeg = get_absolute_value(-c,d)
    while(d <= d_limit):
        while(c <= c_limit):
            if(to_find == absVal):
                print (str(absVal) + ' equals '+ str(to_find))
                print('c = ' + str(c) + ' d = ' + str(d))
                found_pairs.append([c,d])
            if(to_find == absValNeg):
                print (str(absValNeg) + ' equals '+ str(to_find))
                print('c = -' + str(c) + ' d = ' + str(d))
                found_pairs.append([-c,d])
            c += 1
            absVal = get_absolute_value(c,d)
            absValNeg = get_absolute_value(-c,d)
            #print('c = ' + str(c) + ' d = ' + str(d))
        d += 1
        c = 0
        absVal = get_absolute_value(c,d)
        absValNeg = get_absolute_value(-c,d)
    return found_pairs
```

```python
h=7
def f(x, y):
    """ General method that returns the value of the function x² + xy + Cy² or x² + C y², depending on the value of h """
    if h % 4 == 3:
        C = int((h+1)/4)
        return x ** 2 + x * y + C * y ** 2
    else:
        # this covers the case when h equals 1 or 2
        C = h
        return x ** 2 + C * y ** 2

for x in range (0,100):
    for y in range (0,100):
        print((f(x,y)/2)**0.5)
    
```
