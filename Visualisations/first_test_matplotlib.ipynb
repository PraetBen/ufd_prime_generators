```python
import matplotlib.pyplot as plt

points = []

c = []
d = []
for x in range(0,10):
    for y in range(0,10):
        c.append(x+y/2)
        d.append(y*(41**0.5)/2)

plt.scatter(c, d, label="point", marker="d", color="green")
circle1 = plt.Circle((0, 0), 8, color='r')
plt.C
plt.gca().add_patch(circle1)
plt.xlabel('real axis')
plt.ylabel('imaginary axis')
plt.title('complex numbers')
plt.legend()
plt.show()
```
